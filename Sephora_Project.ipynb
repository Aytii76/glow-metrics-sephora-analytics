# **Handle Missing Values**

# Fill small % missing with median

```python
df['rating'] = df['rating'].fillna(df['rating'].median())
df['reviews'] = df['reviews'].fillna(df['reviews'].median())
```

# Fill text-based columns

```python
df['size'] = df['size'].fillna('Unknown')
df['variation_type'] = df['variation_type'].fillna('No Variation')
df['variation_value'] = df['variation_value'].fillna('No Variation')
df['variation_desc'] = df['variation_desc'].fillna('No Variation')
df['ingredients'] = df['ingredients'].fillna('Not Available')
df['highlights'] = df['highlights'].fillna('None')
df['secondary_category'] = df['secondary_category'].fillna('Unknown')
df['tertiary_category'] = df['tertiary_category'].fillna('None')
```

# Fill pricing-related missing values

```python
df['value_price_usd'] = df['value_price_usd'].fillna(0)
df['sale_price_usd'] = df['sale_price_usd'].fillna(0)
df['child_max_price'] = df['child_max_price'].fillna(0)
df['child_min_price'] = df['child_min_price'].fillna(0)
```

```python
# Re-check missing values
df.isnull().sum()
```

```python
df.rename(columns={'price_usd': 'price'}, inplace=True)
```

# Drop unnecessary columns

```python
df.drop(columns=['value_price_usd', 'sale_price_usd', 'limited_edition', 
                 'new', 'online_only', 'out_of_stock', 'sephora_exclusive', 
                 'child_count', 'child_max_price', 'child_min_price'], inplace=True)
```

# Distribution of product prices

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], bins=50, kde=True , color='red')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# Count of products per brand (top 10)

```python
plt.figure(figsize=(12, 6))
top_brands = df['brand_name'].value_counts().head(10)
sns.barplot(x=top_brands.index, y=top_brands.values , color='lightcoral')
plt.title('Top 10 Brands by Product Count')
plt.xticks(rotation=45)
plt.ylabel('Number of Products')
plt.show()
```

# Plot the distribution of ratings

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['rating'], bins=20, kde=True, color='red')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# Top Brands by Number of Reviews

```python
# Group by brand_name and count the number of reviews for each brand
brand_reviews = df.groupby('brand_name')['reviews'].sum().sort_values(ascending=False).head(10)


plt.figure(figsize=(10, 6))
brand_reviews.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Brands by Number of Reviews')
plt.xlabel('Brand')
plt.ylabel('Total Reviews')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#  Top 10 Products by Loves Count

```python
top_loved = df[['product_name', 'loves_count']].sort_values(by='loves_count', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='loves_count', y='product_name', data=top_loved, color='lightcoral')
plt.title('Top 10 Products by Loves Count')
plt.xlabel('Loves Count')
plt.ylabel('Product Name')
plt.show()
```

# Count the number of products in each primary category

```python
category_count = df['primary_category'].value_counts()


plt.figure(figsize=(10, 6))
category_count.plot(kind='bar', color='teal')
plt.title('Product Count by Primary Category')
plt.xlabel('Category')
plt.ylabel('Product Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Box plot to compare price vs rating by primary category

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='primary_category', y='price', data=df, palette='Set3', hue='primary_category')
plt.title('Box Plot for Price vs Rating by Category')
plt.xlabel('Primary Category')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Count the number of products by variation type

```python
variation_type_count = df['variation_type'].value_counts()


plt.figure(figsize=(10, 6))
sns.countplot(y='variation_type', data=df, order=variation_type_count.index, palette='pastel')
plt.title('Product Count by Variation Type')
plt.xlabel('Product Count')
plt.ylabel('Variation Type')
plt.show()
```
